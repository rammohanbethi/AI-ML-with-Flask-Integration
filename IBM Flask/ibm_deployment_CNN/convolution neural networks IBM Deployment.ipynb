```python
import os 
filenames = os.listdir('/home/wsuser/work/Crop-animal data/trainset')
```

```python

x_train = train_datagen.flow_from_directory("/home/wsuser/work/Crop-animal data/trainset",target_size = (64,64),batch_size = 32, class_mode = "categorical")
x_test = test_datagen.flow_from_directory("/home/wsuser/work/Crop-animal data/testset",target_size = (64,64),batch_size = 32, class_mode = "categorical")
```

```python
x_train.class_indices
```

```python
model = Sequential()
```

```python
model.add(Convolution2D(32,(3,3),input_shape = (64,64,3),activation = "relu"))
```

```python
model.add(MaxPooling2D(pool_size = (2,2)))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(units = 128,init = "uniform", activation = "relu"))
```

```python
model.add(Dense(units = 5,init = "uniform", activation = "softmax"))
```

```python
model.compile(optimizer = "adam",loss = "categorical_crossentropy",metrics = ["accuracy"])
```

```python
model.fit_generator(x_train,steps_per_epoch = 47 ,epochs = 10,validation_data = x_test,validation_steps =  20)
```

```python
model.save("animal.h5")
```
